```python
import pandas as pd
```

```python
#1. Creamos un diccionario con la informaci√≥n
datos = {
    'nombre':['Ana','Luis','Carlos'],
    'edad':[30,25,40]
}
```

```python
#2. Convertimos ese diccionario en un DataFrame de Pandas
df_empleados = pd.DataFrame(datos)
```

```python
#3. Mostramos el resultado
print(df_empleados)
```

```python
#Seleccionamos la columna 'edad'
edades = df_empleados['edad']
```

```python
#Mostramos el contenido
print("Contenido de la variable edades:")
print(edades)
```

```python
#Verificamos el tipo de dato
print("\nEl tipo de dato es: ")
print(type(edades))
```
