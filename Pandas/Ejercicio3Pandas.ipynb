```python
import pandas as pd
```

```python
#1. Creamos un diccionario con la información
datos = {
    'nombre':['Ana','Luis','Carlos'],
    'edad':[30,25,40]
}
```

```python
#2. Convertimos ese diccionario en un DataFrame de Pandas
df_empleados = pd.DataFrame(datos)
```

```python
#1. Explorar la forma (filas, columnas)
shape_df = df_empleados.shape
print(shape_df)
```

```python
#2. Explorar los nombres de las columnas
columns_df = df_empleados.columns
print(columns_df)
```

```python
#3. Explorar el índice (las etiquetas de las filas)
index_df = df_empleados.index
print(index_df)
```

```python
#1. Renombrar columnas
#Usamos un diccionario para mapear: 'nombre_viejo': 'nombre_nuevo'
df_empleados = df_empleados.rename(columns={'nombre':'empleado','edad':'años'})
print(df_empleados)
```
